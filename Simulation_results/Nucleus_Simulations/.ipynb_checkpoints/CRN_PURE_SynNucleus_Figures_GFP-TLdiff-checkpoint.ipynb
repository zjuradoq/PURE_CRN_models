## Definitions

```python
def create_custom_plot(title_text, x_max=8,y_max=2, xname='Time (hours)', yname='MGapt (Î¼M)',height=400, width=500):
    custom_plot = figure(
        toolbar_location='right',
        outline_line_color=None,
        min_border_right=10,
        height=height,
        width=width,)

    custom_plot.title.text = title_text
    custom_plot.xaxis.axis_label = xname
    custom_plot.yaxis.axis_label = yname
    custom_plot.y_range = Range1d(0, y_max)
    custom_plot.x_range = Range1d(0, x_max)
    custom_plot.outline_line_color = None

    # custom_plot.yaxis
    custom_plot.ygrid.visible = False
    custom_plot.yaxis.axis_label_text_font_size = '15pt'
    custom_plot.yaxis.major_label_text_font_size = '15pt'
    custom_plot.yaxis.major_label_text_font = 'Work Sans'
    custom_plot.yaxis.axis_label_standoff = 15
    custom_plot.yaxis.axis_label_text_font_style = 'normal'

    # custom_plot.xaxis
    custom_plot.xgrid.visible = False
    custom_plot.xaxis.axis_label_text_font_size = '15pt'
    custom_plot.xaxis.major_label_text_font_size = '15pt'
    custom_plot.xaxis.major_label_text_font = 'Work Sans'
    custom_plot.xaxis.axis_label_standoff = 15
    custom_plot.xaxis.axis_label_text_font_style = 'normal'

    # custom_plot.title
    custom_plot.title.text_font_size = '18pt'
    custom_plot.title.align = "left"
    custom_plot.title.offset=-70

    return custom_plot
```

```python
def updateLegend(plot, title="", location='top_left',visible=True):
    plot.legend.location=location
    plot.legend.click_policy="hide"
    # Remove the box around the legend
    plot.legend.border_line_color = None
    # Add a legend and set its title
    plot.legend.title = title
    plot.legend.title_text_font_style = "normal"
    plot.legend.visible=visible
    plot.legend.label_text_font_size = "12pt"
    # plot.add_layout(plot.legend[0],layout_loc)
    return(plot)
```

# Make Plot

```python
#Plot by k_linker with seperate conditonal tests
i=0
for tl in [25, 50, 90]:  
    count=0

    p_Protein=create_custom_plot('', x_max=3,y_max=7, yname='deGFP (Î¼M)')
    # p_RNA4TL=create_custom_plot('', x_max=3,y_max=3.25, yname='RNA (Î¼M)')
    # p_RNA=create_custom_plot('', x_max=3,y_max=3.25, yname='RNA (Î¼M)')
    
    for expN in [0,-1, -2,-3,-4,-5]:
    # if 'split' in folder:TXnTL_MGapt-deGFP_exp-1TL50
        filename='TXnTL_MGapt-deGFP_exp'+str(expN)+'TL'+str(tl)+'.csv'
        df_data=pd.read_csv(directory+folder+filename)
        df_data[df_data['time'] <= 3*3600]

        # p_RNA.line(df_data['time']/3600, df_data['mRNA_t'], legend_label = f'10^{str(expN)}', line_color=colors[count], line_width=3, line_alpha=.8, line_dash='dashed') #mRNA transcribed  
        p_Protein.line(df_data['time']/3600, df_data['deGFP_m']*.7, legend_label = f'10^{str(expN)}', line_color=colors[count], line_width=3, line_alpha=.8, line_dash=dash[count]) #mRNA transcribed  
        # p_RNA4TL.line(df_data['time']/3600, df_data['mRNA_d'], legend_label = f'10^{str(expN)}', line_color=colors[count], line_width=3, line_alpha=.8, line_dash='dashed') #mRNA transcribed 


        if expN==0:
            PURE_Model_Final=df_data
            Time=np.round(PURE_Model_Final['time']/3600,decimals=1)
            data_consumables= PURE_Model_Final[PURE_Model_Final['time'] <= 3*3600]

            list_ntps=['mRNA_t', 'ATP_tx','GTP_tx', 'CTP_tx','UTP_tx',]
            data_ntps=data_consumables[list_ntps]
            data_ntps['Time']=Time
            data_ntps=data_ntps.set_index('Time')
            data_ntps=np.round(data_ntps/data_ntps.max(),6)
            fig1, ax1 = plt.subplots(figsize=(10,3)) 
            ax1 = sns.heatmap(data_ntps.transpose(),fmt=".1f", #norm=LogNorm(),
                             cbar_kws={'label': 'Normalized Concentration'})

            ax1.set_xlabel("Time (hours)", fontsize=15)
            ax1.set_ylabel('Species', fontsize=15)

            # plt.title("A) NXPs Concentration in Combine PURE Model", loc='left',fontsize=18)
            plt.show()
            image_format = 'svg' # e.g .png, .svg, etc.
            image_name = f'Figures/GFP_TLenergy{tl}_nxps.svg'
            fig1.tight_layout() 

            fig1.savefig(image_name, format=image_format, dpi=1200)

            list_aa= ['deGFP_m','fMettRNAfMetCAU' ,'FD','Gly','Leu','Lys','Asp','Val','Thr','Ala']
            data_aa=data_consumables[list_aa]
            data_aa['Time']=Time
            data_aa=data_aa.set_index('Time')
            data_aa=np.round(data_aa/data_aa.max(),6)
            fig2, ax2 = plt.subplots(figsize=(10,3)) 
            ax2 = sns.heatmap(data_aa.transpose(),fmt=".1f",
                             cbar_kws={'label': 'Normalized Concentration'})

            ax2.set_xlabel("Time (hours)", fontsize=15)
            ax2.set_ylabel('Species', fontsize=15)
            fig2.tight_layout() 

            # plt.title("B) Amino Acids Concentration in Combine PURE Model", loc='left',fontsize=18)
            plt.show()
            image_format = 'svg' # e.g .png, .svg, etc.
            image_name = f'Figures/GFP_TLenergy{tl}_aa.svg'

            fig2.savefig(image_name, format=image_format, dpi=1200)


        count+=1

    #Formatting
    # p_RNA=updateLegend(p_RNA, title="", location='top_left')
    p_Protein=updateLegend(p_Protein, location='top_left')
    # p_RNA4TL=updateLegend(p_RNA4TL, title="", location='top_left')

    p_Protein.output_backend = "svg"
    export_svgs(p_Protein, filename = 'Figures/GFP_25TX'+str(i)+'.svg',width=400, height=300)
    # for plot in [p_Protein]:
        # Output to a file (HTML format)
    # export_png(p_Protein, filename=f"GFP_{fld}{str(i)}.png")
    i+=1
    # # Create a grid layout with two columns
    layout = gridplot([[p_Protein]])
    show(layout)
```
