```python
def create_custom_plot(title_text, x_max=8,y_max=2, xname='Time (hours)', yname='MGapt (M)',height=400, width=500):
    custom_plot = figure(
        toolbar_location='right',
        outline_line_color=None,
        min_border_right=10,
        height=height,
        width=width,)

    custom_plot.title.text = title_text
    custom_plot.xaxis.axis_label = xname
    custom_plot.yaxis.axis_label = yname
    custom_plot.y_range = Range1d(0, y_max)
    custom_plot.x_range = Range1d(0, x_max)
    custom_plot.outline_line_color = None

    # custom_plot.yaxis
    custom_plot.ygrid.visible = False
    custom_plot.yaxis.axis_label_text_font_size = '15pt'
    custom_plot.yaxis.major_label_text_font_size = '15pt'
    custom_plot.yaxis.major_label_text_font = 'Work Sans'
    custom_plot.yaxis.axis_label_standoff = 15
    custom_plot.yaxis.axis_label_text_font_style = 'normal'

    # custom_plot.xaxis
    custom_plot.xgrid.visible = False
    custom_plot.xaxis.axis_label_text_font_size = '15pt'
    custom_plot.xaxis.major_label_text_font_size = '15pt'
    custom_plot.xaxis.major_label_text_font = 'Work Sans'
    custom_plot.xaxis.axis_label_standoff = 15
    custom_plot.xaxis.axis_label_text_font_style = 'normal'

    # custom_plot.title
    custom_plot.title.text_font_size = '18pt'
    custom_plot.title.align = "left"
    custom_plot.title.offset=-70

    return custom_plot
```

```python
def updateLegend(plot, title="", location='top_left',visible=True):
    plot.legend.location=location
    plot.legend.click_policy="hide"
    # Remove the box around the legend
    plot.legend.border_line_color = None
    # Add a legend and set its title
    plot.legend.title = title
    plot.legend.title_text_font_style = "normal"
    plot.legend.visible=visible
    plot.legend.label_text_font_size = "10pt"
    plot.legend.spacing = -3  # Increase the space between legend entries
    return(plot)
```

```python
def tosourcedata(df,ntp):
    # folders=['GFP_klinker', 'GFP_split_TX2th', 'GFP_split_TX4th', 'GFP_split_TX10th',]
    exps=[0,-1,-2,-3,-4,-5]
    sourcedata_df = {}
    sourcedata_df['folder']= ['GFP_shared_CP2','GFP_shared', 'GFP_shared_CPhalf', 
                              'GFP_split_CP2','GFP_split','GFP_split_CPhalf']
    
    # # Iterate over each concentration value
    for exp in df['exp']:
        # Select the row corresponding to the current concentration
        row = df_Energy[df_Energy['exp'] == exp][[ntp+'_tx']].values.tolist()
        # Store the sourcedata for the current concentration in the dictionary
        sourcedata_df[exp] = row
    return sourcedata_df

def barplotsvg(plot,ntp):
    # Set the output backend to SVG
    plot.output_backend = "svg"
    plot.ygrid.grid_line_color = None
    plot.xgrid.grid_line_color = None
    plot.outline_line_color = None

    # Customize the font and size of the x-axis label
    # p.xaxis.axis_label_text_font = "Arial"
    plot.xaxis.axis_label_text_font_size = "12pt"
    plot.xaxis.axis_label_text_font_style ='normal'
    plot.xaxis.axis_label_standoff = 10
    # Customize the font and size of the y-axis label
    # p.yaxis.axis_label_text_font = "Arial"
    plot.yaxis.axis_label_text_font_size = "15pt"
    plot.yaxis.axis_label_text_font_style ='normal'
    plot.yaxis.axis_label_standoff = 10

    # Export the VBar plot as an SVG file
    export_svgs(p, filename=f"Figures/GFP_CPn{ntp}.svg")
    return(plot)
```

```python
def CountString(array, df):
    matching_strings = []
    
    for axp in array:
        substring = axp
        for string in df.columns:
        # Check if the substring is present in the main string
            if substring in string:
            # If present, add the main string to the array
                matching_strings.append(string)
    # print(matching_strings)
    df=df[matching_strings]
    # Sum across columns (axis=1)
    sum_across_columns = df.sum(axis=1)
    EndValue=sum_across_columns.iloc[-1]
    
    return EndValue
    
```

```python
df_Energy=pd.DataFrame(columns=['folder','exp', 'k_linker','mRNA_t','mRNA_d','deGFP_m',
                                'ATP','GTP','CTP','UTP','CP',
                                'ATP_tx', 'GTP_tx', 'CTP_tx', 'UTP_tx','ATP_tl', 'GTP_tl','Pepts'])
j=0
for cp in ['2','1','half']: 
    if cp!='1':
        for expN in [0,-1,-2,-3,-4,-5]:
            filename='/TXTL_MGapt-deGFP_exp'+str(expN)+'wCPx'+cp+'.csv'
            df_data=pd.read_csv(directory+'/'+f_shared+filename)

            df_data[df_data['time'] <= 3*3600]

            #Add information to DF
            data_to_append = pd.DataFrame([[f_shared+cp, str(expN), str(10**expN), ]],
                                          columns=['folder', 'exp','k_linker', ])     
            df_Energy = pd.concat([df_Energy, data_to_append], ignore_index=True)

            for species in ['mRNA_t','mRNA_d','deGFP_m']:
                df_Energy[species][j]=df_data[species][200]

            for species in ['CTP', 'UTP',]:
                df_Energy[species][j]=(df_data[species][0]-df_data[species][200])/1000
                df_Energy[species+'_tx'][j]=df_Energy[species][j]

            for species in ['ATP', 'GTP', 'CP']:
                df_Energy[species][j]=(df_data[species][0]-df_data[species][200])/1000

            for ntp in ['ATP', 'GTP',]:
                if ntp=='ATP':
                    df_Energy[ntp+'_tx'][j]=(df_Energy['mRNA_t'][j]*count_A)/1000
                else:
                    df_Energy[ntp+'_tx'][j]=(df_Energy['mRNA_t'][j]*count_G)/1000

                df_Energy[ntp+'_tl'][j]=df_Energy[ntp][j]-df_Energy[ntp+'_tx'][j]

            for species in ['Pepts']:
                df_Energy[species][j]=(df_Energy['deGFP_m'][j])/CountString(['Pept'],df_data)*100

            j+=1
    else:            
###For just reguakrCP
        f_shared0='GFP_klinker'
        for expN in [0,-1,-2,-3,-4,-5]:

            filename='/TXTL_MGapt-deGFP_exp'+str(expN)+'.csv'
            df_data=pd.read_csv(directory+'/'+f_shared0+filename)

            df_data[df_data['time'] <= 3*3600]

            #Add information to DF
            data_to_append = pd.DataFrame([[f_shared0+cp, str(expN), str(10**expN), ]],
                                          columns=['folder', 'exp','k_linker', ])     
            df_Energy = pd.concat([df_Energy, data_to_append], ignore_index=True)

            for species in ['mRNA_t','mRNA_d','deGFP_m']:
                df_Energy[species][j]=df_data[species][200]

            for species in ['CTP', 'UTP',]:
                df_Energy[species][j]=(df_data[species][0]-df_data[species][200])/1000
                df_Energy[species+'_tx'][j]=df_Energy[species][j]

            for species in ['ATP', 'GTP', 'CP']:
                df_Energy[species][j]=(df_data[species][0]-df_data[species][200])/1000

            for ntp in ['ATP', 'GTP',]:
                if ntp=='ATP':
                    df_Energy[ntp+'_tx'][j]=(df_Energy['mRNA_t'][j]*count_A)/1000
                else:
                    df_Energy[ntp+'_tx'][j]=(df_Energy['mRNA_t'][j]*count_G)/1000

                df_Energy[ntp+'_tl'][j]=df_Energy[ntp][j]-df_Energy[ntp+'_tx'][j]

            for species in ['Pepts']:
                df_Energy[species][j]=(df_Energy['deGFP_m'][j])/CountString(['Pept'],df_data)*100

            j+=1
        
#For split system
for cp in ['2','1','half']: 
    if cp!='1':
        for expN in [0,-1,-2,-3,-4,-5]:
            filename='/TXnTL_MGapt-deGFP_exp'+str(expN)+'wCPx'+cp+'.csv'
            df_data=pd.read_csv(directory+f_split0+f_split+filename)
            df_data[df_data['time'] <= 3*3600]

            #Add information to DF
            data_to_append = pd.DataFrame([[f_split+cp, str(expN), str(10**expN), ]],
                                          columns=['folder', 'exp','k_linker', ])     
            df_Energy = pd.concat([df_Energy, data_to_append], ignore_index=True)

            for species in ['mRNA_t','mRNA_d','deGFP_m']:
                df_Energy[species][j]=df_data[species][200]

            for species in ['ATP_tx', 'GTP_tx', 'CTP_tx', 'UTP_tx','CP']:
                df_Energy[species][j]=(df_data[species][0]-df_data[species][200])/1000

            for species in ['ATP', 'GTP',]:
                df_Energy[species+'_tl'][j]=(df_data[species][0]-df_data[species][200])/1000
                df_Energy[species][j]=df_Energy[species+'_tl'][j]+df_Energy[species+'_tx'][j]

            for species in ['CTP', 'UTP',]:
                df_Energy[species][j]=df_Energy[species+'_tx'][j]

            for species in ['Pepts']:
                df_Energy[species][j]=(df_Energy['deGFP_m'][j])/CountString(['Pept'],df_data)*100

            j+=1

    else:
        # ###For just reguakrCP
        f_split_4th='GFP_split_TX4th' # [2]
        #For split system
        for expN in [0,-1,-2,-3,-4,-5]:

            filename='/TXnTL_MGapt-deGFP_exp'+str(expN)+'.csv'
            df_data=pd.read_csv(directory+f_split0+f_split_4th+filename)
            df_data[df_data['time'] <= 3*3600]


            #Add information to DF
            data_to_append = pd.DataFrame([['GFP_split'+cp, str(expN), str(10**expN), ]],
                                          columns=['folder', 'exp','k_linker', ])     
            df_Energy = pd.concat([df_Energy, data_to_append], ignore_index=True)

            for species in ['mRNA_t','mRNA_d','deGFP_m']:
                df_Energy[species][j]=df_data[species][200]

            for species in ['ATP_tx', 'GTP_tx', 'CTP_tx', 'UTP_tx','CP']:
                df_Energy[species][j]=(df_data[species][0]-df_data[species][200])/1000

            for species in ['ATP', 'GTP',]:
                df_Energy[species+'_tl'][j]=(df_data[species][0]-df_data[species][200])/1000
                df_Energy[species][j]=df_Energy[species+'_tl'][j]+df_Energy[species+'_tx'][j]

            for species in ['CTP', 'UTP',]:
                df_Energy[species][j]=df_Energy[species+'_tx'][j]

            for species in ['Pepts']:
                df_Energy[species][j]=(df_Energy['deGFP_m'][j])/CountString(['Pept'],df_data)*100

            j+=1
```
